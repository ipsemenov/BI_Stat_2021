#### 3. Way

There is a lot of code here but the idea is rather simple: we shuffle columns in the data, calculate PCA and obtain explained variance ratio for each component on the random data. Then we repeat this procedure many times to obtain some distribution and for our initial data we choose only those components which explaine more variance than components with he same numbers but  calculated for the random data.

```python
def shuffle_dataset(dataset):
    """
    Shuffle data
    """
    random_data =  {}
    for col in dataset.columns:
        random_data[col] = np.random.permutation(dataset.loc[:, col].values)
    random_data = pd.DataFrame(random_data)
    
    return random_data


def get_variance_by_chance(dataset, n_replics, n_components):
    """
    Calculate expalined variance ratio for each component on the shuffled data
    """
    variance_explained_by_chance = np.zeros((n_replics, n_components))
    
    for i in tqdm.tqdm_notebook(range(n_replics)):
        if i % 100 == 0:
            print(f"{i} iter")

        random_data = shuffle_dataset(dataset)
        random_pca = PCA(n_components=n_components)
        random_pca.fit(random_data)
        variance_explained_by_chance[i, :] = random_pca.explained_variance_ratio_
        
    return variance_explained_by_chance


def get_pc_variance(dataset, n_components):
    """
    Calculate expalined variance ratio for each component on the initial data
    """
    pca = PCA(n_components=n_components)
    pca.fit(dataset)
    
    return pca.explained_variance_ratio_

# There are some graphs for visualization

def plot_mean_and_ci(ax, values, label, ci_level=0.95, alpha_transparency=0.5, color_mean=None, color_shading=None):
    mean = values.mean(axis=0)
    
    std = values.std(axis=0)
    n = values.shape[1]
    se = std / np.sqrt(n) 
    
    q_alpha = (1-ci_level) / 2
    ci_num = np.abs(norm.ppf(q_alpha, loc=0, scale=1))
    
    lb = mean -  ci_num * se
    ub = mean + ci_num * se
    
    ax.fill_between(range(mean.shape[0]), ub, lb,
                     color=color_shading, alpha=alpha_transparency)
    
    ax.plot(mean, c=color_mean, lw=3, label=label)
    
    
def plot_explained_variance(ax, variance):
    ax.plot(variance, label="real", lw=3)
    ax.scatter(np.arange(0, variance.shape[0]), variance)
    
    
def plot_variance_by_change(ax, variance_by_chance):
    plot_mean_and_ci(ax, variance_by_chance, label="chance", color_mean="red", color_shading="red")


def plot_explained_vs_chance(ax, explained_variance, variance_by_chance, dataset_name, step=1):
    plot_explained_variance(ax, explained_variance)
    plot_variance_by_change(ax, variance_by_chance)

    ax.set_title(f"PCA {dataset_name}", size=35)
    ax.set_xlabel("Component number", size=25)
    ax.set_ylabel("Explained variance ratio", size=25)
    ax.set_xticks(np.arange(0, explained_variance.shape[0], step))
    ax.set_xticklabels(np.arange(1, explained_variance.shape[0]+1, step), size=20)

    ax.tick_params(labelsize=20, size=10)
    ax.set_ylim(0, explained_variance[0] + 0.1 )
    ax.legend(fontsize=20)

    
def plot_pval_plot(ax, p_values, dataset_name, alpha_level = 0.05, logscale=True, step=1):
    if logscale:
        p_values = -np.log10(p_values)
        alpha_level = -np.log10(alpha_level)

    
    ax.set_title(f"PC significance, {dataset_name}", size=35)
    ax.plot(p_values, lw=3)
    ax.scatter(np.arange(0, p_values.shape[0]), p_values, lw=3)

    ax.set_xlabel("Component number", size=25)
    ax.set_ylabel("-log(pvalue + eps)", size=25)
    ax.set_xticks(np.arange(0, p_values.shape[0], step))
    
    ax.set_xticklabels(labels = np.arange(1, p_values.shape[0]+1, step), size=20)
    ax.tick_params(labelsize=20, size=10)
    

    ax.hlines(y=alpha_level, xmin=0, xmax=p_values.shape[0], color="red", linestyles="dashed", lw=3)


def calc_permutat_pval(real_values, permut_values, eps=None):
    """
    Compare explained variance ratio for initial and shuffled data
    """
    eps = eps or (1 / (permut_values.shape[0] * 10))
    
    p_values = np.zeros_like(real_values)
    for i in range(0, p_values.shape[0], 1):
        p_values[i] = (permut_values[:, i] >= real_values[i]).mean() + eps
    return p_values

    
def pca_analysis(ax1, ax2, dataset, title, n_components = None, n_replics = 1000, step=1):
    """
    The main function which runs all above code
    """
    n_components = n_components or dataset.shape[1]
    explained_variance = get_pc_variance(dataset, n_components)
    variance_by_chance = get_variance_by_chance(dataset, n_replics, n_components)
    p_values = calc_permutat_pval(explained_variance, variance_by_chance)
    plot_explained_vs_chance(ax1, explained_variance, variance_by_chance, title)
    plot_pval_plot(ax2, p_values, title)
```

```python
f, (ax1, ax2) = plt.subplots(2, 1)
f.set_figheight(20)
f.set_figwidth(20)
pca_analysis(ax1, ax2, X_scRNAseq, "sc-RNAseq", n_replics=10, n_components=15)
```

### t-SNE

```python
X_reduced = PCA(n_components=47).fit_transform(X_scaled)
init = PCA(n_components=2).fit_transform(X_scaled)
n_clusters = len(Y_scRNAseq.unique())
```

```python
def plot_tsne(data, n_clusters , perplexity_ls, n_iter=5000, metric='cosine', init='random', n_cols=3):
    n_rows = int(np.ceil(len(perplexity_ls) / n_cols))
    time_start = time.time()
    plt.figure(figsize=(20, 20));
    for i, perplexity in enumerate(perplexity_ls):
        tsne = TSNE(n_components=2, perplexity=perplexity, 
                    n_iter=n_iter, metric=metric, init=init)
        tsne_results = tsne.fit_transform(data)
        tsne_df = pd.DataFrame()
        tsne_df['y'] = Y_scRNAseq
        tsne_df['t-SNE 1'] = tsne_results[:,0]
        tsne_df['t-SNE 2'] = tsne_results[:,1]
        plt.subplot(n_rows, n_cols, i+1);
        sns.scatterplot(x="t-SNE 1", y="t-SNE 2",
                        hue="y", palette=sns.color_palette("hls", n_clusters),
                        data=tsne_df, legend=False, alpha=0.3);
        plt.xlabel('t-SNE 1', fontsize=12);
        plt.ylabel('t-SNE 2', fontsize=12);
        plt.title(f'Perplexity: {perplexity}, Step: {n_iter}', fontsize=14)

    total_time = str(datetime.timedelta(seconds=time.time() - time_start))
    print(f't-SNE done! Time elapsed: {total_time}')
```

**t-SNE on reduced data after PCA (47 components kept)**

```python
plot_tsne(data=X_reduced, n_clusters=n_clusters, perplexity_ls=[5, 10, 20, 30, 40, 50, 100])
```

**t-SNE on all initial scaled data with initialization with PCA (first 2 components)**

```python
plot_tsne(data=X_scaled, n_clusters=n_clusters, perplexity_ls=[5, 10, 20, 30, 40, 50, 100], init=init)
```

**t-SNE on reduced data after PCA (47 components kept) with initialization with PCA (first 2 components)**

```python
plot_tsne(data=X_reduced, n_clusters=n_clusters, perplexity_ls=[5, 10, 20, 30, 40, 50, 100], init=init)
```

**Conсlusions**:
* Making t-SNE on data reduced by PCA let us to get rid of noise in our data which is observed at low perplexity.
* Initialization with PCA (first 2 components) makes t-SNE faster

### UMAP

```python
def plot_umap(data, n_clusters, parameters, metric='cosine', n_cols=4):
    time_start = time.time()
    n_rows = int(np.ceil(len(parameters) / n_cols))
    plt.figure(figsize=(20, 35));
    for i, pair in enumerate(parameters):
        n_neighbors, min_dist = pair
        umap = UMAP(n_components=2, n_neighbors=n_neighbors, 
                    min_dist=min_dist, metric=metric)
        umap_results = umap.fit_transform(X_reduced)
        umap_df = pd.DataFrame()
        umap_df['y'] = Y_scRNAseq
        umap_df['UMAP 1'] = umap_results[:,0]
        umap_df['UMAP 2'] = umap_results[:,1]
        plt.subplot(n_rows, n_cols, i+1);
        sns.scatterplot(x="UMAP 1", y="UMAP 2",
                        hue="y", palette=sns.color_palette("hls", n_clusters),
                        data=umap_df, legend=False, alpha=0.3);
        plt.xlabel('');
        plt.ylabel('');
        plt.title(f'n_neighbors: {n_neighbors}, min_dist: {min_dist}', fontsize=14)
    total_time = str(datetime.timedelta(seconds=time.time() - time_start))
    print(f'UMAP done! Time elapsed: {total_time}')
```

```python
n_clusters = len(Y_scRNAseq.unique())
n_neighbors_ls = [5, 10, 20, 50, 100]
min_dist_ls = [0.0, 0.1, 0.5, 0.8, 0.99]
parameters = list(product(n_neighbors_ls, min_dist_ls))
```

```python
plot_umap(data=X_reduced, n_clusters=n_clusters, parameters=parameters)
```

**Conclusions**:
* UMAP is much faster than t-SNE
* But with set parameters it seems that t-SNE better separates clusters
